```python
import fiftyone
import fiftyone.utils.openimages

dataset_train = fiftyone.zoo.load_zoo_dataset(
              "open-images-v7",
              split="train",
              label_types=["detections"],
              classes=["Car", "Vehicle registration plate"],
              max_samples=1500,
              only_matching=True,
          )
```

```python
import fiftyone
import fiftyone.utils.openimages

dataset_val = fiftyone.zoo.load_zoo_dataset(
              "open-images-v7",
              split="validation",
              label_types=["detections"],
              classes=["Car", "Vehicle registration plate"],
              max_samples=300,
              only_matching=True,
          )
```

```python
!mkdir dataset
!mkdir dataset/labels
!mkdir dataset/images
!mkdir dataset/labels/train
!mkdir dataset/labels/val
!mkdir dataset/images/train
!mkdir dataset/images/val
```

```python
import shutil

def to_YOLO_format(dataset, split):
    count_car = 0
    count_license_plate = 0
    for sample in dataset:
        isGroupOf = False
        for detection in sample.ground_truth.detections:
            if detection.IsGroupOf != 0:
                isGroupOf = True
                break
        if isGroupOf == True:
            continue
        image_name = sample.filepath[sample.filepath.rfind("/")+1:]
        shutil.copyfile(f"{sample.filepath}", f"dataset/images/{split}/{image_name}")
        image_name = image_name[:-4]
        with open(f"dataset/labels/{split}/{image_name}.txt", "w") as file:
            for detection in sample.ground_truth.detections:
                if "Car" in detection.label:
                    file.write("0 ")
                    count_car += 1
                else:
                    file.write("1 ")
                    count_license_plate += 1
                bounding_box = detection.bounding_box
                file.write(str(bounding_box[0]+bounding_box[2]/2) + " "
                           + str(bounding_box[1]+bounding_box[3]/2) + " "
                           + str(bounding_box[2]) + " " + str(bounding_box[3]) + "\n")
    return count_car, count_license_plate
```

```python
train_count = to_YOLO_format(dataset_train, "train")
val_count = to_YOLO_format(dataset_val, "val")
print(f"In train dataset there are {train_count[0]} cars and {train_count[1]} license plates")
print(f"In validation dataset there are {val_count[0]} cars and {val_count[1]} license plates")
```

```python
with open("dataset/data.yaml", "w") as file:
    file.write('''
path: ../dataset
train: images/train
val: images/val

names:
  0: car
  1: license_plate
''')
```
